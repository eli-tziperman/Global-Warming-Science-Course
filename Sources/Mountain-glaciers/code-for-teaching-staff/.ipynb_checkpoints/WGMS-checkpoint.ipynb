# Mountain glaciers:

```python
# load libraries, read time series data:
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import griddata
from numpy import linalg
from scipy import stats
import cartopy.crs as ccrs
from scipy.stats import linregress
from scipy.signal import savgol_filter # for smoothing time series
from scipy.interpolate import griddata
from scipy import optimize
import copy
import time
from datetime import datetime

# uncomment for interactive plots, following https://stackoverflow.com/questions/50149562/jupyterlab-interactive-plot
#%matplotlib widget

# font required by PUP:
plt.rcParams['font.family'] = 'Myriad Pro'
fontsize=9
plt.rcParams['font.size'] = fontsize

def year_or_none(d):
    if not isinstance(d, str):
        return None
    try:
        return datetime.strptime(d, "%Y-%m-%d").year
    except ValueError:
        return None

# Import the csv datasets:
# -----------------------
# File with Glacier data front location:
filename='../../../Data-for-teaching-staff/Mountain-glaciers/WGMS/DOI-WGMS-FoG-2025-02b/data/' \
   +'front_variation.csv.gz'
dataset = pd.read_csv(filename, encoding = "latin",sep=',',na_values='NaN' \
              , low_memory=False, comment='#', skipinitialspace=False \
              , skip_blank_lines=False, header=0, compression="gzip")
data = np.asarray(dataset.iloc[:, :].values)
header = dataset.iloc[:, :].columns
all_political_units=data[:,0]
all_glacier_names=data[:,1]
all_years=np.array([year_or_none(d) for d in data[:,4]], dtype=float)
all_front_variation=data[:,8]
print("header:",header)
print("len(all_years)=",len(all_years),"; years:",all_years)
print("glaciers names:",all_glacier_names)

# File with general Glacier data, including lon/lat:
filename='../../../Data-for-teaching-staff/Mountain-glaciers/WGMS/DOI-WGMS-FoG-2025-02b/data/glacier.csv.gz'
dataset1 = pd.read_csv(filename, encoding = "latin",sep=',',na_values='NaN',low_memory=False \
                       ,comment='#',skipinitialspace=False,skip_blank_lines=False,header=0, compression="gzip")
data1 = dataset1.iloc[:, :].values
header1 = dataset1.iloc[:, :].columns
all_political_units1=data1[:,0]
all_glacier_names1=data1[:,1]
all_lat1=pd.to_numeric(data1[:,4], errors="coerce").astype(float)
all_lon1=pd.to_numeric(data1[:,5], errors="coerce").astype(float)
print("header1:",header1)
print("lat:",all_lat1)
print("lon:",all_lon1)

# first year for plots and last year to include in the analysis:
start_year=1850
last_year_with_data=2024
print("start_year=", start_year, ", last_year_with_data=", last_year_with_data)
```

```python
# divide the data into separate records for each glacier, start with edge locations::

# ---------------------------------------------------------------------------------------------------------
# ---------------------------------------------------------------------------------------------------------
# front_variation{} dictionary includes: 
# {glacier name: [years],[front location],[lon,lat]
#               ,usable,[regression_a,regression_b]}
# (initially the first entry is front variation, later changed below to front location.)
# ---------------------------------------------------------------------------------------------------------
# ---------------------------------------------------------------------------------------------------------

front_variation = {}
for i in range(len(all_years)):
    glacier_name = all_glacier_names[i] + '/' + all_political_units[i]

    y  = all_years[i]
    dv = all_front_variation[i]

    # --- better fix: skip rows with missing year or missing length_change ---
    if not (np.isfinite(y) and np.isfinite(dv)):
        continue

    if glacier_name in front_variation:
        front_variation[glacier_name][0].append(y)
        front_variation[glacier_name][1].append(dv)
    else:
        front_variation[glacier_name] = [[y], [dv], [360.0, 90.0], True, []]

# front_variation={}
# for i in range(len(all_years)):
#     glacier_name=all_glacier_names[i]+'/'+all_political_units[i]
#     if glacier_name in front_variation.keys():
#         # if glacier already in data, add to its dictionary values:
#         if not np.isnan(all_front_variation[i]):
#             # turns out same glacier name may appear in different countries (political units), 
#             # so include political unit in key denoting which glacier it is:
#             front_variation[glacier_name][0].append(all_years[i])
#             front_variation[glacier_name][1].append(all_front_variation[i])
#     else:
#         # if glacier is not in dictionary, add it first, and then add to its values:
#         if True: # not np.isnan(all_front_variation[i]):
#             front_variation.update({glacier_name : 
#                                 [[all_years[i]],[all_front_variation[i]],[360.0,90.0],True,[]]})

# use the other dataset to add lon/lat to the front_variation dictionary:
for i in range(len(all_glacier_names1)):
    glacier_name1=all_glacier_names1[i]+'/'+all_political_units1[i]
    if glacier_name1 in front_variation.keys():
        #print("adding (lon,lat) to glacier ",glacier_name1)
        lon=all_lon1[i]
        lat=all_lat1[i]
        front_variation[glacier_name1][2]=[lon,lat]

print("done.")
```

```python
# specify plot/analysis parameters:
reference_year=1960
min_length=10 # minimum number of front location for a given glacier
max_front_variation=6000 # max absolute value total variation in front location to include in analysis:

Nrecord_not_usable=0
Nrecords_usable=0
i=-1
for glacier_name in front_variation.keys():
    i=i+1
    # interpolate to find value at reference_year:
    years=np.asarray(front_variation[glacier_name][0])
    # do not include records that start after reference year, or end before 
    # reference year, or are too long, 
    # or show very large variation, or years are not monotonically increasing:
    if years[0]>reference_year or years[-1]<reference_year \
        or len(front_variation[glacier_name][1])<min_length:
        #or not np.all(np.diff(front_variation[glacier_name][0]) > 0):
        #or np.any(np.isnan(front_variation[glacier_name][2])): 
        Nrecord_not_usable=Nrecord_not_usable+1
        front_variation[glacier_name][3]=False # set "usable" property in dictionary to False
        if 0: # print additional information for debugging:
            print("*** record for glacier",glacier_name," starts only at year ",years[0]
                  ,"and has a length=",len(front_variation[glacier_name][1]))
    else:
        locations=np.cumsum(front_variation[glacier_name][1])
        location_at_reference_year=griddata(years, locations, reference_year, method='linear')
        # remove value at reference_year so all series are zero then:
        locations=np.asarray(locations-location_at_reference_year)
        # put locations rather than location changes into the dictionary:
        front_variation[glacier_name][1]=locations
        front_variation[glacier_name][3]=True # "usable"
        if np.amax(locations)-np.amin(locations)>max_front_variation:
            front_variation[glacier_name][3]=False
            Nrecord_not_usable=Nrecord_not_usable+1
        else:
            Nrecords_usable=Nrecords_usable+1


print("Analyzing",Nrecords_usable,"records. Records for",Nrecord_not_usable,"out of",i
      ,"glaciers \n   started after or ended before reference year"\
      ,reference_year,"\n   or had too short record, or too large variability, \n   or non-monotonic years, or had no lat/lon info"\
      ,"\n   and are ignored.")        

print("done.")
```

```python
# prepare data to be picked for students by removing all nonusable records:

front_variation_to_pickle=copy.deepcopy(front_variation)
for key, value in front_variation.items():
    if value[3]==False:
        del(front_variation_to_pickle[key])
front_variation=copy.deepcopy(front_variation_to_pickle)

# now remove the items used to store usability and regression coefficients only 
# from the versiont to be pickled:        
for key, value in front_variation_to_pickle.items():
    del(front_variation_to_pickle[key][3:])

np.save("Output/to-pickle/front_variation.npy",front_variation_to_pickle)
print("done.")
```

```python
# plot glacier location time series:

fig=plt.figure(figsize=(6,4),dpi=200)
markers=['.','+','o','v','^','<','>','8','P','p','s','*','h','H','x','X','D','d',4,5,6,7,8,9,10,11]
i=-1
for glacier_name in front_variation.keys():
    i=i+1
    years=np.asarray(front_variation[glacier_name][0])
    locations=np.asarray(front_variation[glacier_name][1])
    # plot:
    plt.plot(years,locations/1000,'-',lw=0.5,marker=markers[np.mod(i,len(markers))]
        ,markersize=1.5,label=glacier_name)

plt.xlabel("year")
plt.xlim(1850,2025)
#plt.ylim(-5000,2200)
plt.ylabel("front location relative to 1960 (km)")
plt.grid(lw=0.25)

plt.tight_layout()
plt.show()
```

```python
# Follow "Extracting a Climate Signal from 169 Glacier Records" 
# by J. Oerlemans and calculate average length relative to a reference year:
# calculate the bin-averaged record by average over decadal-bins

props = dict(boxstyle='round', edgecolor="white", facecolor='white', alpha=1.0)

# first, combine all glacier record into a single one:
years_combined =[]; locations_combined=[];
for glacier_name in front_variation.keys():
        years=front_variation[glacier_name][0]
        locations=front_variation[glacier_name][1]
        # remove value at start_year so all series start at the same plate
        years_combined.extend(years)
        locations_combined.extend(locations)


# bin-averaged glacier length:
bin_length_in_years=3
bin_edges=np.arange(1600,2040,bin_length_in_years)
bin_means, bin_edges, binnumber \
      = stats.binned_statistic(years_combined,locations_combined, statistic='mean',bins=bin_edges)
# bin-averaged std:
bin_std, bin_edges, binnumber \
      = stats.binned_statistic(years_combined,locations_combined, statistic='std',bins=bin_edges)
# number of data points per bin:
bin_count, bin_edges, binnumber \
      = stats.binned_statistic(years_combined,locations_combined, statistic='count',bins=bin_edges)
years_bins_axis=(bin_edges[0:-1]+bin_edges[1:])*0.5


# calculate years of data per bin, given the last_year_with_data:
years_with_data_per_bin=bin_count*0+bin_length_in_years
years_with_data_per_bin[np.logical_and(bin_count<1,years_bins_axis>=start_year)]=np.nan
first_nan_index = np.argmax(np.isnan(years_with_data_per_bin))
years_with_data_per_bin[first_nan_index-1]=last_year_with_data-bin_edges[first_nan_index-1]+1


# load instrumental GISS global-mean annual-mean temperature record:
GISS_obs=np.load("../../Temperature/code-for-teaching-staff/Output/to-pickle/OBS_GISS_GMST_annual_anomaly_timeseries.npy")
# smooth GISS temperature:
GISS_obs_smooth=1.0*GISS_obs
GISS_obs_smooth[1] = savgol_filter(GISS_obs[1], 1, 0) # window size , polynomial order 


# plot:
fig=plt.figure(figsize=(7,2.5),dpi=300)
plt.subplot(1,2,1)
plt.plot(years_bins_axis,bin_means/1000,drawstyle='steps-mid',lw=1,color="b")
plt.xlabel("Year")
plt.ylabel("Front location (km)") # ,color="b")
plt.xlim(1850,last_year_with_data)
plt.ylim(-1000/1000,1600/1000)
plt.grid(lw=0.25)
ax1=plt.gca()
ax1.text(0.09, 0.55, "(a)", transform=ax1.transAxes, fontsize=12,
        horizontalalignment='center',verticalalignment='center', bbox=props)


plt.subplot(1,2,2)
plt.semilogy(years_bins_axis, bin_count/years_with_data_per_bin,drawstyle='steps-mid',lw=1,color="b")
plt.xlabel("Year")
plt.ylabel("Number of observations/year")
plt.xlim(1850,last_year_with_data)
plt.ylim(1,4000)
plt.grid(lw=0.25)
axes=plt.gca()
axes.text(0.09, 0.55, "(b)", transform=axes.transAxes, fontsize=12,
        horizontalalignment='center',verticalalignment='center', bbox=props)

plt.tight_layout()
plt.show()
```

```python
# calculate linear regression for each glacier:
for glacier_name in front_variation.keys():
        years=front_variation[glacier_name][0]
        locations=front_variation[glacier_name][1]
        # calculate linear regression and store result in dictionary:
        fit=linregress(years, locations)
        slope=fit.slope;
        intercept=fit.intercept;
        p=fit.pvalue;
        r2=(fit.rvalue)**2
        front_variation[glacier_name][4]=[intercept,slope]

print("done.")
```

## Calculate temperature from Glacier extent

Follow Oerleman's calculation of temperature from glacier extent,
$$\frac{dL'(t)}{dt}=-\frac{1}{\tau}[cT'(t)+L'(t)]$$
which implies that
$$T'(t)=-\frac{1}{c}[L'(t)+\tau\frac{dL'(t)}{dt}]$$


```python
# read Hadley surface temperature since 1850 and calculate annual average:
# ------------------------------------------------------------------------

dataset = pd.read_csv('../../../Data-for-teaching-staff/Temperature/Hadley-SST/HadCRUT5.0Analysis_gl.txt'\
                      ,sep=r'\s+',na_values='NaN',comment='#',skipinitialspace=True,skip_blank_lines=True,header=None)
data = np.asarray(dataset.iloc[:, :].values)
#header = dataset.iloc[:, :].columns
# exclude the last year, as glacier data does not extend far enough:
Ts_years=1.0*data[::2,0] # ::2 skips the lines with error bars etc
Ts_annual=1.0*data[::2,-1]
# dont use the last year which is incomplete:
Ts_annual=Ts_annual[Ts_years<=last_year_with_data]
Ts_years=Ts_years[Ts_years<=last_year_with_data]

# interpolate binned data to annual glacier location time series:
# ---------------------------------------------------------------
glacier_years=1.0*years_bins_axis[np.logical_and(years_bins_axis>=1850,years_bins_axis<=last_year_with_data)]-0.5
glacier_locations=1.0*bin_means[np.logical_and(years_bins_axis>=1850,years_bins_axis<=last_year_with_data)]
glacier_years1=np.arange(1850,last_year_with_data,1)
glacier_locations1=griddata(glacier_years, glacier_locations, glacier_years1, method='linear')


# smooth time series:
# -------------------
# smooth anual mean time series:
Ts_annual_smooth = savgol_filter(Ts_annual, 7, 1) # window size , polynomial order 
# set Ts to be anomaly relative to 1960:
Ts_annual_smooth=Ts_annual_smooth-Ts_annual_smooth[Ts_years==1960]

# smooth anual mean glaceir location time series:
glacier_locations_smooth1 = savgol_filter(glacier_locations1[:], 15, 1) # window size , polynomial order 

# use Oerleman's formula to get temperature from glacier extent:
# --------------------------------------------------------------
glacier_locations_smooth2=1.0*glacier_locations_smooth1[glacier_years1>=start_year]
glacier_years2=1.0*glacier_years1[glacier_years1>=start_year]
Ts_annual_smooth2=1.0*Ts_annual_smooth[Ts_years>=start_year]


def cost(x):
    # cost for minimizing distance between reconstructed and observed temperature
    c=x[0]
    tau=x[1]
    T0=0 #x[2]
    T_glacier=T0+(-1/c)*(glacier_locations_smooth2[1:-1]+tau*dLdt)
    J=np.sum(np.abs(T_glacier-Ts_annual_smooth2[1:-2])**2)
    return J

# initial guess for minimization:
c=2000 # m/deg
tau=20 # years
x=np.zeros(2)
x[0]=c
x[1]=tau
#x[2]=0 # offset, not in Oerlemans' formula
# dL/dt used in cost evaluation:
dt=2 # years, for center difference
dLdt=(glacier_locations_smooth2[2:]-glacier_locations_smooth2[:-2])/dt
# minimize:
sol = optimize.minimize(cost, x)
c=sol.x[0]
tau=sol.x[1]
T0=0 #sol.x[2]
print("optimal solution is c=",c,"tau=",tau,"T0=",T0)

# calculate reconstructed temperature using optimal parameters:
T_glacier=T0+(-1/c)*(glacier_locations_smooth2[1:-1]+tau*dLdt)

c_equib=c
T_glacier_equilib=(-1/c_equib)*(glacier_locations_smooth2)


# -----
# plot:
# -----
fig=plt.figure(figsize=(4.9,4.15),dpi=300)
plt.subplot(2,2,1)
plt.plot(Ts_years,Ts_annual,lw=0.5,color='r')
plt.plot(Ts_years,Ts_annual_smooth,lw=2,color='b')
plt.xlabel("Year")
plt.ylabel("$T'_{obs}$ (°C)")
plt.xlim(start_year,last_year_with_data)
plt.ylim(-0.6,1.0)
plt.yticks(np.arange(-0.5,1.05,0.25))
plt.grid(lw=0.25)
plt.title("(a)", loc="left")

plt.subplot(2,2,2)
plt.plot(glacier_years1,glacier_locations1,lw=0.5,color='r')
plt.plot(glacier_years2,glacier_locations_smooth2,lw=1,color='b')
plt.xlim(start_year,last_year_with_data)
plt.xlabel("Year")
plt.ylabel("Front location (m)",labelpad=0)
plt.grid(lw=0.25)
plt.title("(b)", loc="left")

plt.subplot(2,1,2)
plt.plot(Ts_years[1:-1],Ts_annual_smooth[1:-1],lw=2,color="b",label="Observed")
plt.plot(glacier_years2[1:-1],T_glacier,color="r",lw=2,label="Reconstructed")
plt.plot(glacier_years2,T_glacier_equilib,color="r",ls="--",lw=1,label="Equilibrium")
plt.ylabel("Temperature anomaly (°C)")
plt.xlabel("Year")
plt.xlim(start_year,last_year_with_data)
plt.yticks(np.arange(-0.5,1.0,0.25))
plt.grid(lw=0.25)
plt.title("(c)", loc="left")
plt.legend();

plt.tight_layout(pad=0)
plt.subplots_adjust(hspace=0.35,wspace=0.4)
fig.savefig("Output/mountain-glaciers-Oerlemans-optimization.pdf"
           ,bbox_inches='tight',pad_inches = 0)


np.save("Output/to-pickle/Ts_smooth.npy",Ts_annual_smooth)
np.save("Output/to-pickle/Ts_smooth_years.npy",Ts_years)
```
