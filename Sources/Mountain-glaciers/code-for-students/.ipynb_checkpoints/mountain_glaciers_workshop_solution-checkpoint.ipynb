## 3. Idealized glacier length adjustment scenarios:


#### (a) Plot time series of idealized glacier length adjustment scenarios with a temperature that changes abruptly at t=0 and then remains constant, and using the exponential solution to equation (11.1). Use the same c and τ parameters calculated in question 2. For the first scenario, let the initially imposed temperature anomaly $T_0'$ be $−1$ °C and the initial length $L_0'$ be 4000 m. For the second scenario, $T_0'=−2$ °C and $L_0'=1000$ m.

```python
# set time array and various parameters for both scenarios:
t=np.asarray(np.arange(0,100,1))
tau=15
c=1800
T0a=-1
T0b=-2
L0a=4000
L0b=1000

# The solutions for both scenarios:
La=(L0a+c*T0a)*np.exp(-t/tau)-c*T0a
Lb=(L0b+c*T0b)*np.exp(-t/tau)-c*T0b
label_a="$T_0'="+repr(T0a)+r"{\rm °C},\ L_0'=$"+repr(L0a/1000)+" km"
label_b="$T_0'="+repr(T0b)+r"{\rm °C},\ L_0'=$"+repr(L0b/1000)+" km"

# plot both solutions:
fig=plt.figure(figsize=(5,3),dpi=300)
plt.rcParams.update({"text.usetex": False})
plt.plot(t,La/1000,color="r",label=label_a)
plt.plot(t,Lb/1000,color="b",label=label_b)
plt.xlabel("Time (years)")
plt.ylabel("Lengtn anomaly, $L'(t)$ (km)")
plt.xlim(-1,100)
plt.grid(lw=0.25)
plt.legend()
plt.tight_layout();
fig.savefig("Output/mountain-glaciers-idealized-adjustment.pdf"
           ,bbox_inches='tight',pad_inches = 0)
```

#### (b) Optional extra credit: Glacier length response to a varying temperature: Find and plot the solution to equation (11.1) for a linearly changing temperature. Use start and end anomaly temperatures of $T_0=−0.25$ K, $T_f=1$ K, climate sensitivity of $c=3000$ m/K, adjustment time of $τ=20$ yr, a total period of 140 yr, and an initial length anomaly of $L_0=6000$ m. Explain the structure of the solution you find.

Solution: The equation is
\begin{align}
  \frac{dL'(t)}{dt}&=-\frac{1}{\tau}\left( L'(t)+c(T_0+at) \right)\\
                   &=-AL+Bt+C
\end{align}
where $a=(T_f-T_0)/\Delta t$, the initial and final temperatures are $T_0,T_f$, the total time period is given by $\Delta t$, $A=1/\tau$, $B=-ac/\tau$ and $C=-cT_0/\tau$. The solution (https://www.wolframalpha.com/input/?i=dL%2Fdt%3D-A*L%2BB*t%2BC) is 
\begin{align}
    L(t)&=-B/A^2+Bt/A+c_1e^{-At}+C/A\\
        &=(ac/\tau)\tau^2 - (ac/\tau)t\tau+c_1 e^{-t/\tau} - cT_0\\
        &=ca(\tau - t)+c_1 e^{-t/\tau} + -cT_0\\
\end{align}
so that,
\begin{align}
    L(t)&=c\frac{T_f-T_0}{\Delta t}(\tau-t) + c_1 e^{-t/\tau} - cT_0.
\end{align}
Now use an initial condition of $L(t=0)=L_0=1.3$ km to find $c_1$. 
\begin{align}
    L_0&=c\frac{T_f-T_0}{\Delta t}\tau + c_1 -cT_0\\
        c_1&=-c\frac{T_f-T_0}{\Delta t}\tau + L_0 +cT_0.
\end{align}

```python
# set parameters:
L0=6000 # m
c=3000 # m/K
tau=20 # years
T0=-0.25 # K
Tf=1     # K
Deltat=140 # years
c1=-c*tau*(Tf-T0)/Deltat+L0+c*T0
time=np.arange(0,Deltat,Deltat/100)
T=T0+((Tf-T0)/Deltat)*time
L=c*((Tf-T0)/Deltat)*(tau-time)+c1*np.exp(-time/tau)-c*T0

# plot prescribed linear temperature:
plt.figure(figsize=(7,3),dpi=200)
plt.subplot(1,2,1)
plt.plot(time,T)
plt.xlabel("years")
plt.ylabel("Temperature")
plt.grid(lw=0.25)

# plot resulting glacier length anomaly record:
plt.subplot(1,2,2)
plt.plot(time,L/1000)
plt.xlabel("years")
plt.ylabel("lengtn anomaly, $L'(t)$ (km)")
plt.grid(lw=0.25)

plt.tight_layout()
print("initial L=",L[0])
```

## 5) Optional extra credit: The global picture: Calculate a linear regression for each of the given glacier records from the global data set. Draw a global map with circles indicating the latitude and longitude of glaciers, with red indicating negative length trends (slope). Make the size of the circles reflect the magnitude of the trend. Superimpose blue symbols at the location of glaciers showing positive trends.

```python
# calculate linear regression for each glacier:
slopes = []
for glacier_name in front_variation.keys():
    years = np.asarray(front_variation[glacier_name][0], dtype=float)
    locations = np.asarray(front_variation[glacier_name][1], dtype=float)

    m = np.isfinite(years) & np.isfinite(locations)
    if m.sum() < 2:
        continue  # not enough data to fit

    fit = linregress(years[m], locations[m])
    slope = fit.slope
    intercept = fit.intercept
    slopes.append(slope)

    if len(front_variation[glacier_name]) < 4:
        front_variation[glacier_name].append([intercept, slope])
    else:
        front_variation[glacier_name][3] = [intercept, slope]

# calculate standard deviation for normalizing symbol size based on 
# trend magnitude:
std_slope=np.std(slopes)

# replace trend with normalized one for plotting purposes:
for glacier_name in front_variation.keys():
    normalized_slope=front_variation[glacier_name][3][1]/std_slope
    normalized_slope=np.minimum(2,normalized_slope)
    normalized_slope=np.maximum(-2,normalized_slope)
    front_variation[glacier_name][3][1]=normalized_slope
    
# mark glacier locations on a map, indicating separately those with 
# positive and negative trends:

# draw a map with coastlines first:
projection=ccrs.PlateCarree(central_longitude=0.0);
fig,axes=plt.subplots(1,1,figsize=(6,4.5),dpi=300\
                      ,subplot_kw={'projection': projection});
axes.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())
axes.coastlines(resolution='110m',lw=0.3,color="grey")
axes.gridlines(linewidth=0.2)
#axes.set_title('location of WGMS front location data')
gl = axes.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=0.5, color='gray', alpha=0.5, linestyle='-')
gl.left_labels = True
gl.right_labels = False
gl.top_labels = False


# mark a red/blue markes at glacier locations depending if they are 
# decreasing/ increasing based on the linear regression:
# negative trends
N_negative_trends=0
for glacier_name in front_variation.keys():
        lon,lat=front_variation[glacier_name][2]
        if front_variation[glacier_name][3][1]<0:
                N_negative_trends=N_negative_trends+1
                axes.plot(lon,lat,'o',color='r',markerfacecolor='none'\
                        ,markeredgewidth=0.5 \
                        ,markersize=-3*front_variation[glacier_name][3][1]
                        ,transform=ccrs.PlateCarree())

# positive trends:
N_positive_trends=0
for glacier_name in front_variation.keys():
        lon,lat=front_variation[glacier_name][2]
        if front_variation[glacier_name][3][1]>=0:
            N_positive_trends=N_positive_trends+1
            axes.plot(lon,lat,'o',color='b'\
                     ,markeredgewidth=0.5 \
                     ,markersize=3*front_variation[glacier_name][3][1]
                     ,transform=ccrs.PlateCarree())

plt.show()

print("found",N_negative_trends,"negative trends and",N_positive_trends\
      ,"positive trends.")
fig.savefig("Output/mountain-glaciers-map-with-trend-signs.pdf"
           ,bbox_inches='tight',pad_inches = 0)
```

## (6) Plot Northern Hemisphere snow area time series for each season.

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6), sharey=False, dpi=300)
axes = axes.flatten()

start_date = pd.Timestamp('1980-01-01')
end_date = pd.Timestamp('2024-12-31')
view_start = pd.Timestamp('1980-01-01')
view_end   = pd.Timestamp('2025-01-01') # Extends to 2025 for the tick mark

seasons = ['DJF', 'MAM', 'JJA', 'SON']
season_names = {
    'DJF': 'Winter (Dec-Feb)',
    'MAM': 'Spring (Mar-May)',
    'JJA': 'Summer (Jun-Aug)',
    'SON': 'Autumn (Sep-Nov)'
}

for i, season in enumerate(seasons):
    ax = axes[i]
    
    # extract data from dictionary:
    time = snow_extent_north_hemisphere[season]["time"]
    area = snow_extent_north_hemisphere[season]["area"]
        
    # Create Decimal Year for Regression (e.g., 2000.5)
    decimal_year = time.dt.year + time.dt.dayofyear / 365.25
    
    # Linear Regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(decimal_year, area)
    
    # Generate the regression line
    line = slope * decimal_year + intercept
    
    # Plotting
    ax.plot(time, area, label='Snow Extent', color='steelblue', linewidth=1.5)
    ax.plot(time, line, label=f'Trend: ${slope*10:.2f}$ M km$^2/$dec', color='red', linestyle='--')

    # Formatting
    ax.set_title(season_names[season], fontsize=12)#, fontweight='bold')
    ax.set_ylabel('Snow extent ($10^6$ km$^2$)')
    if i==2 or i==3:
        ax.set_xlabel('Year')        
    ax.set_xlim(view_start, view_end)
    ax.grid(True, linestyle=':', alpha=0.6)
    ax.legend(loc='best')

    # 2. Major Ticks (The labels with numbers)
    # We set this to every 5 years so the text doesn't overlap, 
    # but it ensures 1980, 1985... 2020, 2025 are hit.
    ax.xaxis.set_major_locator(mdates.YearLocator(10))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    
    # 3. Minor Ticks (The small ticks without numbers)
    # Set this to 1 year to satisfy your "every year" requirement
    ax.xaxis.set_minor_locator(mdates.YearLocator(1))

plt.tight_layout()
## save as pdf:
f = plt.gcf()
f.savefig("Output/mountain-glaciers-snow-extent-NH.pdf",format='pdf');
```
