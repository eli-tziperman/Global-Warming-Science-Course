```python
# load libraries, read time series data from the local Tableau packaged workbook (.twbx):
import io
import os
import sys
import zipfile
import shutil
import tempfile
import subprocess
from pathlib import Path
from datetime import datetime, timedelta

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy import linalg
import pickle

TWBX_PATH = Path.home() / "Downloads" / "3 - Landing Page Fires by area and number.twbx"
# Fallback for this chat environment / uploaded file location.
if not TWBX_PATH.exists():
    alt_path = Path("/mnt/data/3 - Landing Page Fires by area and number.twbx")
    if alt_path.exists():
        TWBX_PATH = alt_path


def ensure_tableauhyperapi():
    try:
        from tableauhyperapi import HyperProcess, Telemetry, Connection  # noqa: F401
        return
    except ModuleNotFoundError:
        print("Installing tableauhyperapi ...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "tableauhyperapi"])


def _name_to_str(name):
    if hasattr(name, "unescaped"):
        return str(name.unescaped)
    return str(name)


def _table_name_sql(table_name):
    # Tableau Hyper API returns table names as objects such as TableName(schema, table).
    # Build a safely quoted SQL identifier without depending on one exact object layout.
    if hasattr(table_name, "schema_name") and hasattr(table_name, "table_name"):
        schema = _name_to_str(table_name.schema_name)
        table = _name_to_str(table_name.table_name)
        return f'"{schema.replace(chr(34), chr(34) * 2)}"."{table.replace(chr(34), chr(34) * 2)}"'
    txt = str(table_name)
    if "." in txt:
        return txt
    return f'"{txt.replace(chr(34), chr(34) * 2)}"'


def list_hyper_tables(connection):
    catalog = connection.catalog
    tables = []
    schema_names = catalog.get_schema_names()
    for schema_name in schema_names:
        try:
            table_names = catalog.get_table_names(schema_name)
        except TypeError:
            table_names = catalog.get_table_names(schema=schema_name)
        for table_name in table_names:
            tables.append(table_name)
    return tables


def list_columns(connection, table_name):
    table_def = connection.catalog.get_table_definition(table_name)
    cols = []
    for col in table_def.columns:
        cols.append(_name_to_str(col.name))
    return cols


def find_matching_columns(cols, kind):
    norm = {c.casefold(): c for c in cols}

    year_candidates = [
        "year",
        "année",
    ]
    if kind == "area":
        value_candidates = [
            "area (hectares)",
            "superficie (en hectare)",
        ]
    elif kind == "number":
        value_candidates = [
            "number",
            "nombre",
        ]
    else:
        raise ValueError(f"Unsupported kind: {kind}")

    year_col = next((norm[c] for c in year_candidates if c in norm), None)
    value_col = next((norm[c] for c in value_candidates if c in norm), None)
    return year_col, value_col


def extract_series_from_hyper(hyper_path, kind):
    from tableauhyperapi import HyperProcess, Telemetry, Connection

    with HyperProcess(Telemetry.DO_NOT_SEND_USAGE_DATA_TO_TABLEAU) as hyper:
        with Connection(endpoint=hyper.endpoint, database=str(hyper_path)) as connection:
            debug = []
            for table_name in list_hyper_tables(connection):
                cols = list_columns(connection, table_name)
                year_col, value_col = find_matching_columns(cols, kind)
                debug.append(f"table={table_name} cols={cols}")
                if year_col is None or value_col is None:
                    continue

                table_sql = _table_name_sql(table_name)
                year_sql = '"' + year_col.replace('"', '""') + '"'
                value_sql = '"' + value_col.replace('"', '""') + '"'
                sql = f"""
                    SELECT CAST({year_sql} AS INTEGER) AS year,
                           SUM(CAST({value_sql} AS DOUBLE)) AS value
                    FROM {table_sql}
                    GROUP BY 1
                    ORDER BY 1
                """
                rows = connection.execute_list_query(sql)
                if not rows:
                    continue
                years = np.asarray([int(row[0]) for row in rows], dtype=int)
                values = np.asarray([float(row[1]) for row in rows], dtype=float)
                mask = np.isfinite(years) & np.isfinite(values)
                years = years[mask]
                values = values[mask]
                if years.size == 0:
                    continue
                return years, values

    raise ValueError(
        f"Could not find a usable {kind!r} table in {hyper_path}. "
        f"Available tables/columns: {' | '.join(debug)}"
    )


ensure_tableauhyperapi()

with tempfile.TemporaryDirectory() as tmpdir:
    tmpdir = Path(tmpdir)
    with zipfile.ZipFile(TWBX_PATH) as zf:
        zf.extractall(tmpdir)

    hyper_files = sorted(tmpdir.rglob("*.hyper"))
    if len(hyper_files) < 2:
        raise ValueError(f"Expected at least two .hyper files inside {TWBX_PATH}, found {len(hyper_files)}")

    Canada_fire_area_years = None
    Canada_fire_area = None
    Canada_fire_number_years = None
    Canada_fire_number = None
    area_errors = []
    number_errors = []

    for hyper_file in hyper_files:
        try:
            years, values = extract_series_from_hyper(hyper_file, "area")
            Canada_fire_area_years = years
            Canada_fire_area = values / 100.0  # hectares -> km^2
            break
        except Exception as exc:
            area_errors.append(f"{hyper_file.name}: {exc}")

    for hyper_file in hyper_files:
        try:
            years, values = extract_series_from_hyper(hyper_file, "number")
            Canada_fire_number_years = years
            Canada_fire_number = values
            break
        except Exception as exc:
            number_errors.append(f"{hyper_file.name}: {exc}")

    if Canada_fire_area is None:
        raise ValueError("Could not extract area-burned series from the .twbx file.\n" + "\n".join(area_errors))
    if Canada_fire_number is None:
        raise ValueError("Could not extract number-of-fires series from the .twbx file.\n" + "\n".join(number_errors))

```

```python
# plot:
plt.figure(dpi=600)
plt.subplot(2,1,1)
plt.plot(Canada_fire_area_years,Canada_fire_area)
plt.xlabel("year")
plt.ylabel("area burnt (km$^2$)")
plt.grid(lw=0.25)

plt.subplot(2,1,2)
plt.plot(Canada_fire_number_years,Canada_fire_number)
plt.xlabel("year")
plt.ylabel("number of fires")
plt.grid(lw=0.25)

plt.tight_layout()
```

```python
np.save("Output/to-pickle/canada_fire_years",Canada_fire_area_years)
np.save("Output/to-pickle/canada_fire_area",Canada_fire_area)
np.save("Output/to-pickle/canada_fire_number",Canada_fire_number)
print("done.")
```
