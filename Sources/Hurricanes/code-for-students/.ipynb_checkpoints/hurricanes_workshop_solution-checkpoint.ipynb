#### (c) Smooth the two time series using two passes of a 1-2-1 filter ("running average"),
\begin{align*}
      \overline{T}_i=0.25T_{i-1}+0.5T_{i}+0.25T_{i+1}.
\end{align*}
#### Note that every pass requires eliminating the 1st and last data points, where the smoothing cannot be evaluated. Write the expression for the twice-smoothed temperature $\overline{\overline{T}}_i$ in terms of $T_i$. Over how many years is the average done in this case? 

#### Plot the smoothed time series as thick lines over the un-smoothed ones shown as thin lines.

#### Plot the PDI and the SST averaged over the Atlantic Main Development Area (MDR), on the same axes (using two separate $y$ axes).

```python
        
def smooth_time_series(data):
    """Apply one pass of 1-2-1 filter to input data.
    Inputs: data - a time series
    Output: smoothed data, two less in length then input, as first 
    and last cannot be calculated
    """
    # one pass of smoothing:
    data_smooth = np.zeros(len(data))
    for j in range(1, len(data)-1):
        data_smooth[j]=.25*data[j-1] + 0.5*data[j] + 0.25*data[j+1]
    # return smoohted data without first and last values:
    return data_smooth[1:len(data)-1]


# calculate smoothed time series (2 passes) using above function:
# ----------------------------------------------------------------
# For the SST:
# first pass:
SST_smooth=smooth_time_series(SST_MDR_OBS)
SST_smooth_years=1.0*SST_MDR_OBS_years[1:len(SST_MDR_OBS_years)-1]
# second pass:
SST_smooth=smooth_time_series(SST_smooth)
SST_smooth_years=1.0*SST_smooth_years[1:len(SST_smooth_years)-1]
# first pass for the PDI:
PDI_smooth=smooth_time_series(PDI)
PDI_smooth_years=1.0*PDI_years[1:len(PDI_years)-1]
# second pass:
PDI_smooth=smooth_time_series(PDI_smooth)
PDI_smooth_years=1.0*PDI_smooth_years[1:len(PDI_smooth_years)-1]


# plot the two time series on same time axis, using two vertical axes:
# --------------------------------------------------------------------
fig, axis_PDI = plt.subplots(figsize=(5,3),dpi=300)
# add another y-axis for plotting PDI and SST on the same graph:
axis_SST = axis_PDI.twinx()
# y-axis colors:
axis_SST.yaxis.label.set_color("blue")
axis_SST.tick_params(axis='y', colors='blue')
axis_PDI.yaxis.label.set_color("red")
axis_PDI.tick_params(axis='y', colors='red')
# labels:
axis_PDI.set_ylabel("Annual unscaled PDI")
axis_SST.set_ylabel("SST (°C)")
axis_PDI.set_xlabel("Year")
# axis ranges:
axis_SST.set_ylim(26.0,28.5)
plt.xlim(1850,2025)

# plot the PDI on one y axis and the SST on another
hl1=axis_SST.plot(SST_MDR_OBS_years, SST_MDR_OBS, color="blue"
                  , linewidth=.25, label="SST")
hl2=axis_SST.plot(SST_smooth_years, SST_smooth, linewidth=1.5
                  , label="smoothed", color="blue")
hl3=axis_PDI.plot(PDI_years, PDI, color="red", linewidth=.25
                  , label="PDI")
hl4=axis_PDI.plot(PDI_smooth_years, PDI_smooth, linewidth=1.5
                  , label="smoothed", color="red")

# show legend and display figure:
lines = hl1 + hl2 + hl3 + hl4
labels = [l.get_label() for l in lines]
plt.legend(lines,labels,ncol=2)
plt.tight_layout()
plt.show()
fig.savefig("Output/hurricanes-PDI-and-SST-timeseries.pdf")
```

## Smoothing details
given $\overline{T}_i=0.25T_{i-1}+0.5T_{i}+0.25T_{i+1}$, we have
\begin{align*}
\overline{\overline{T}}_i
&=0.25\overline{T}_{i-1}+0.5\overline{T}_{i}+0.25\overline{T}_{i+1}\\
&=0.25(0.25T_{i-2}+0.5T_{i-1}+0.25T_{i})+0.5(0.25T_{i-1}+0.5T_{i}+0.25T_{i+1})
+0.25(0.25T_{i}+0.5T_{i+1}+0.25T_{i+2}) \\
&=0.0625T_{i-2}+0.25T_{i-1}+0.375T_{i}+0.25T_{i+1}+0.0625T_{i+2}
\end{align*}
so that two passes are equivalent to a 5-point running average.

#### (d) Correlation of SST and PDI: Calculate the correlation between the time series of the SST and PDI with and without smoothing for 1950 to 2005. Repeat using the data up to present-day. To calculate the correlation of the two time series, first remove their mean and then calculate the correlation by programming the explicit formula for correlation. Check your results using the NumPy function for calculating correlation. How would you justify calculating the correlation for the twice-smoothed data?

```python
# method 1: using a built-in numpy function
# call the function for unsmoothed values
builtin_r2_unsmooth = np.corrcoef(PDI, SST_MDR_OBS)[0, 1]**2
builtin_r2_unsmooth_recent = np.corrcoef(PDI[PDI_years>1949] \
                    , SST_MDR_OBS[SST_MDR_OBS_years>1949])[0, 1]**2
print("built-in r² of PDI/SST (unsmoothed):", builtin_r2_unsmooth)
print("built-in r² of PDI/SST (unsmoothed, only 1950-today):" \
      , builtin_r2_unsmooth_recent)


# call the function for smoothed values
builtin_r2 = np.corrcoef(PDI_smooth, SST_smooth)[0, 1]**2
print("built-in r² of PDI/SST (smoothed):  ", builtin_r2)


# calculate correlation for the years 1950-2005:
PDI_smooth_short=PDI_smooth[np.logical_and(PDI_smooth_years>1949 \
                                           ,PDI_smooth_years<2006)]
PDI_smooth_short_years=PDI_smooth_years[np.logical_and(PDI_smooth_years>1949 \
                                                       ,PDI_smooth_years<2006)]
SST_smooth_short=SST_smooth[np.logical_and(SST_smooth_years>1949 \
                                           ,SST_smooth_years<2006)]
builtin_r2_smooth_short = np.corrcoef(PDI_smooth_short \
                                      , SST_smooth_short)[0, 1]**2
print("built-in r² of PDI/SST (smoothed, 1950-2005 only):  " \
      , builtin_r2_smooth_short)

# calculate correlation for the years 1950-today:
PDI_smooth_recent=PDI_smooth[PDI_smooth_years>1949]
PDI_smooth_recent_years=PDI_smooth_years[PDI_smooth_years>1949]
SST_smooth_recent=SST_smooth[SST_smooth_years>1949]
builtin_r2_smooth_recent = np.corrcoef(PDI_smooth_recent \
                                       , SST_smooth_recent)[0, 1]**2
print("built-in r² of PDI/SST (smoothed, 1950-today only):  " \
      , builtin_r2_smooth_recent)


# method 2: an explicit calculation of correlation
def my_calc_correlation(X,Y):
    # remove mean from X:
    X=X-np.mean(X)
    # remove mean from Y:
    Y=Y-np.mean(Y)
    C_XY=np.sum(X*Y)/(np.sqrt( np.sum(X*X)*np.sum(Y*Y) ))
    return C_XY

print("\nVerifying results using my function:")
# call my_calc_correlation for unsmoothed values
my_r2 = my_calc_correlation(PDI, SST_MDR_OBS)
print("my r² of PDI/SST (unsmoothed):", my_r2**2)
# call my_calc_correlation for smoothed values
my_r2 = my_calc_correlation(PDI_smooth, SST_smooth)
print("my r² of PDI/SST (smoothed):  ", my_r2**2)
```

## 2. Potential intensity:


#### (a) Plot the saturation specific humidity as a function of time based on the observed MDR SST and the SST increase projected by 2100 under the RCP8.5 scenario.

```python
def q_sat(T,P):
    # saturation specific humidity (g water vapor per g moist air):
    # inputs:
    # T: temperature, in Kelvin
    # P: pressure, in mb

    R_v = 461 # Gas constant for moist air = 461 J/(kg*K)
    R_d = 287 # Gas constant 287 J K^-1 kg^-1
    TT = T-273.15 # Kelvin to Celsius
    # Saturation water vapor pressure (mb) from Emanuel 4.4.14 p 116-117: 
    ew = 6.112*np.exp((17.67 * TT) / (TT + 243.5))
    # saturation mixing ratio (g water vapor per g dry air):
    rw = (R_d / R_v) * ew / (P - ew)
    # saturation specific humidity (g water vapor per g moist air):
    qw = rw / (1 + rw)
    return qw


# calculate for plotting saturation moisture as function of temperature:
T_plot=np.arange(-15.0,40.0,1.0)+273.15
q_plot=T_plot*0.0
q_plot_approx=T_plot*0.0
for i in range(0,len(T_plot)):
    q_plot[i]=q_sat(T_plot[i],1000)
    q_plot_approx[i]=2.541e6*np.exp(-5415/T_plot[i])*18/29


# calculate and plot saturation moisture for observed and projected MDR SST:
qsat=SST_MDR_OBS*0
P_surface=1000
for i in range(0,len(SST_MDR_OBS)):
    qsat[i]=q_sat(SST_MDR_OBS[i]+273.15,P_surface)

qsat_RCP85=SST_MDR_RCP85*0
for i in range(0,len(SST_MDR_RCP85)):
    qsat_RCP85[i]=q_sat(SST_MDR_RCP85[i]+273.15,P_surface)

f1=plt.figure(1,figsize=(8,8/3),dpi=300)
props = dict(boxstyle='round', edgecolor="wheat", facecolor='wheat', alpha=0.9)
ax=plt.subplot(1,3,1)
plt.plot(T_plot,q_plot*1000,color="blue",lw=1,label="exact")
plt.ylabel("$q^*$ (gr/kg)")
plt.xlabel("$T$ (K)")
plt.xlim([260,310])
plt.ylim([0,40])
plt.title("(a)",loc="left")
# ticks: major every 50, minor every 10
ax.xaxis.set_major_locator(mticker.MultipleLocator(10))
ax.xaxis.set_minor_locator(mticker.MultipleLocator(1))
plt.grid(lw=0.25)

ax=plt.subplot(1,3,2)
plt.plot(SST_MDR_OBS_years,SST_MDR_OBS,color="blue",lw=1,label="Historical")
plt.plot(SST_MDR_RCP85_years,SST_MDR_RCP85,color="red",lw=1,label="RCP8.5")
plt.ylabel("SST (°C)")
plt.xlabel("Year")
plt.xlim(1850,2100)
plt.legend()
plt.title("(b)",loc="left")
# ticks: major every 50, minor every 10
ax.xaxis.set_major_locator(mticker.MultipleLocator(50))
ax.xaxis.set_minor_locator(mticker.MultipleLocator(10))
plt.grid(lw=0.25)

ax=plt.subplot(1,3,3)
plt.plot(SST_MDR_OBS_years,qsat*1000,color="blue",lw=1,label="Historical")
plt.plot(SST_MDR_RCP85_years,qsat_RCP85*1000,color="red",lw=1,label="RCP8.5")
plt.ylabel("$q^*$ (gr/kg)")
plt.xlabel("Year")
plt.xlim(1850,2100)
plt.legend()
plt.title("(c)",loc="left")
# ticks: major every 50, minor every 10
ax.xaxis.set_major_locator(mticker.MultipleLocator(50))
ax.xaxis.set_minor_locator(mticker.MultipleLocator(10))
plt.grid(lw=0.25)

plt.tight_layout(w_pad=0.0)
plt.show()
f1.savefig("Output/hurricanes-Clausius-Clapeyron-SST-qsat-historical-and-RCP.pdf" \
           ,format='pdf');


# for notes:
# ----------
f2=plt.figure(2,figsize=(4,4),dpi=300)
plt.plot(T_plot,q_plot_approx*1000,color="blue",ls="-")
plt.ylabel("$q^*$ (gr/kg)")
plt.xlabel("$T$ (K)");
plt.grid(lw=0.25)
plt.tight_layout(w_pad=0.0)

#f2.savefig("Output/hurricanes-Clausius-Clapeyron.pdf",format='pdf');

T=30+273.15;
print("T=",T,"q*=",2.541e9*np.exp(-5415/T)*18/29);
T=31+273.15;
print("T=",T,"q*=",2.541e9*np.exp(-5415/T)*18/29);
```

#### (b) Calculate and plot the expected potential intensity as a function of time for the observed (historical) MDR SST and for the projected MDR SST increase by year 2100 under the RCP8.5 scenario (plotting the raw PI, with the smoothed time series superimposed).

```python
def calc_potential_intensity_wind_speed_from_SST(sst):
    """Calculate the potential intensity given a sea surface temperature.
    Input: sst- Sea surface temperature, in degrees Celsius
    Output: Return a velocity in m/s
    """
    sst = sst + 273.15 # convert to Kelvin
    epsilon = (1/3)    # efficiency of Hurricane as a carnot engine
    L = 2260*1000      # latent heat of vaporization of water
    PI=sst*0           # initialize array of potential intensity
    for i in range(0,len(sst)):
        q = q_sat(sst[i], 1000) # assumes that pressure is always 1000mb (sea level)
        PI[i] = np.sqrt(epsilon*L*0.15*q)
    return PI


# calculate Potential Intensity (PI) for observed atlantic MDR SST hurricane 
# season record:
PI_SST_MDR_hurricane_season = calc_potential_intensity_wind_speed_from_SST(SST_MDR_OBS)
PI_SST_MDR_hurricane_season_years=SST_MDR_OBS_years
PI_SST_MDR_hurricane_season_smooth=smooth_time_series(PI_SST_MDR_hurricane_season)
PI_SST_MDR_hurricane_season_smooth_years=SST_MDR_OBS_years[1:len(SST_MDR_OBS_years)-1]

# calculate PI for RCP8.5 atlantic MDR SST hurricane season record:
PI_rcp_SST_MDR_hurricane_season = \
       calc_potential_intensity_wind_speed_from_SST(SST_MDR_RCP85)
PI_rcp_SST_MDR_hurricane_season_years=SST_MDR_RCP85_years
 
PI_rcp_SST_MDR_hurricane_season_smooth= \
       smooth_time_series(PI_rcp_SST_MDR_hurricane_season)
PI_rcp_SST_MDR_hurricane_season_smooth_years= \
       SST_MDR_RCP85_years[1:len(SST_MDR_RCP85_years)-1]


# plot smoothed Huricane season PI and SST:
# ------------------------------------------
fig1, ax1 = plt.subplots(dpi=300)
plt.ylabel("Potential intensity (m/s)")
plt.xlabel("Year")
#plt.title("Hurricane season Potential Intensity")
plt.plot(PI_SST_MDR_hurricane_season_smooth_years \
         ,PI_SST_MDR_hurricane_season_smooth
         ,label="PI (smoothed historical SST)",lw=2)
plt.plot(PI_rcp_SST_MDR_hurricane_season_smooth_years \
         ,PI_rcp_SST_MDR_hurricane_season_smooth
         ,label="PI (smoothed RCP8.5 SST)",lw=2)

# plot un-smoothed MDR potential intensity:
# -----------------------------------------
plt.plot(PI_SST_MDR_hurricane_season_years, PI_SST_MDR_hurricane_season, "g-"
         ,label="PI (historical SST)",lw=0.25)
plt.plot(PI_rcp_SST_MDR_hurricane_season_years, PI_rcp_SST_MDR_hurricane_season
         ,color="coral", label="PI (RCP8.5 SST)",lw=0.25)
plt.grid(lw=0.25)
plt.xlim(1850,2100)
plt.legend()
plt.show()
fig1.savefig("Output/hurricanes-PI-historical-and-RCP.pdf")
```

## 3. Optional extra credit: Detecting ACC in hurricane intensity: Calculate the fraction of major hurricanes every year and plot it as a function of time. Repeat after averaging the time series into groups of three years. Calculate and plot a linear fit with and without the three-year averages, and calculate the $r^2$ in each case. Discuss your results.

```python
# plot yearly data without any bin-averages:
# ------------------------------------------

# Create dictionary where keys are the years and the value is an array
# containing the number of major hurricanes and the number of hurricanes
HURSAT_years_axis=np.arange(1978,2018,1)
Hurricanes_per_year = {}
for yr in np.arange(HURSAT_years_axis[0],HURSAT_years_axis[-1]+1):
    Hurricanes_per_year[yr] = [0, 0]

# Loop through every windspeed of every hurricane. If greater than 100
# knots, increase major hurricane and total hurricane count by 1, 
# else if greater than 65 knots, increase only total hurricane count by 1
for ind, hurricane_6hourly_winds in enumerate(HURSAT_wind_speed):
    year = int(HURSAT_years[ind][0])

    # Exclude years 1978 and 1980 
    # (following Kossin et al 2020, due to data quality issues during these years):
    if year != 1978 and year != 1980:
            for windspeed in hurricane_6hourly_winds:
                if windspeed >= 100:
                    # add 1 to both number of hurricanes and number of major hurricanes
                    Hurricanes_per_year[year][0] += 1
                    Hurricanes_per_year[year][1] += 1
                else:
                    if windspeed >= 65:
                        # add 1 to number of hurricanes:
                        Hurricanes_per_year[year][1] += 1
                        
    else: # set both number of hurricanes and number of major hurricanes to 0:
        Hurricanes_per_year[year][0] = 0
        Hurricanes_per_year[year][1] = 0

fraction_major_hurricanes_per_year = []
for key, value in Hurricanes_per_year.items():
    if value[1] != 0:
        fraction_major_hurricanes_per_year.append(value[0]/value[1])
    else:
        fraction_major_hurricanes_per_year.append(np.nan)

        
# calculate linear regresion line:
x1=1.0*HURSAT_years_axis[~np.isnan(fraction_major_hurricanes_per_year)]
y1=np.array(fraction_major_hurricanes_per_year)
y1=y1[~np.isnan(y1)]
slope, intercept, r_value, p_value, std_err =  stats.linregress(x1, y1)
y=intercept+slope*HURSAT_years_axis
print("linear regression diagnostics:")
print("Slope: {}".format(slope))
print("Intercept: {}".format(intercept))
print("R-squared: {}".format(r_value**2))
print("p-value: {}".format(p_value))


# Plot:
# -----
fig=plt.figure(figsize=(5,3),dpi=300)

plt.scatter(x1, y1)
plt.plot(x1, y1, linewidth=1)
plt.xlabel ("Year")
plt.ylabel ("Fraction major hurricanes")
plt.xlim(1977,2020)

# Plot the linear fit
plt.plot(HURSAT_years_axis, y, c='r', linewidth=2)
plt.grid(lw=0.25)

plt.tight_layout()
fig.savefig("Output/hurricanes-fraction-major-hurricanes-timeseries.pdf")
```

```python
# now plot the 3-year bin-averages:
# ----------------------------------
# (Please follow Kossin et al 2020: ignore 1978 and 1980 due to data quality issues;  
#   your first 3-year average is therefore over 1979, 1980 (no data) and 1981)

N=len(fraction_major_hurricanes_per_year)
HURSAT_years_axis_3yr_bins=np.arange(HURSAT_years_axis[0]+2,HURSAT_years_axis[-1],3)
N_3yr_bins=len(HURSAT_years_axis_3yr_bins)
fraction_major_hurricanes_per_year_3yr_bins=np.zeros(N_3yr_bins)*np.nan
ibin=0
for i in range(1,N,3):
    fraction_major_hurricanes_per_year_3yr_bins[ibin]= \
       np.nanmean(fraction_major_hurricanes_per_year[i:i+3]) 
    ibin=ibin+1

    
# Plot:
# -----
fig=plt.figure(figsize=(5,3),dpi=300)
plt.scatter(HURSAT_years_axis_3yr_bins, fraction_major_hurricanes_per_year_3yr_bins)
plt.plot(HURSAT_years_axis_3yr_bins, fraction_major_hurricanes_per_year_3yr_bins, linewidth=1)
plt.xlabel ("Year")
plt.ylabel ("Fraction major hurricanes")
plt.xlim(1977,2020)

# calculate linear regresion line:
x1=1.0*HURSAT_years_axis_3yr_bins
y1=np.array(fraction_major_hurricanes_per_year_3yr_bins)
y1=y1[~np.isnan(y1)]
slope, intercept, r_value, p_value, std_err =  stats.linregress(x1, y1)
y_3yr_bins=intercept+slope*HURSAT_years_axis_3yr_bins
print("linear regression diagnostics:")
print("Slope: {}".format(slope))
print("Intercept: {}".format(intercept))
print("R-squared: {}".format(r_value**2))
print("p-value: {}".format(p_value))

# Plot the linear fit
plt.plot(HURSAT_years_axis_3yr_bins, y_3yr_bins, c='r', linewidth=2)
plt.grid(lw=0.25)

plt.tight_layout()
fig.savefig("Output/hurricanes-fraction-major-hurricanes-timeseries-3yr-bin.pdf")
```
