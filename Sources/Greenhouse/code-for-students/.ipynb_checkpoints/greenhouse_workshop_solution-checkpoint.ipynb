##  2)  Energy balance model: 

(a) Calculate  the downwelling LW and absorbed SW at the surface in the two-layer model, using an albedo of 0.25 and an atmospheric emissivity of 0.8.

```python
# constants:
S0 = 1361
albedo = 0.25
epsilon = 0.8
sigma = 5.6696e-8


# calculate atmospheric temperature T_a:
T = (((S0/4)*((1-albedo)))/(sigma*(1-epsilon/2)))**(1/4)
T_a=T/2**0.25
print("T,T_a=",T,T_a)
# answer:
LW_down=epsilon*sigma*T_a**4
SW_net=S0/4*(1-albedo)
print("LW_down=",LW_down,",SW_net=",SW_net)
```

(b) Modify the 2-layer energy balance model to take into account the absorption of a fraction $\Delta=0.15$ of the downward shortwave radiation by the atmosphere. Calculate the resulting temperatures, compare to the case of $\Delta=0$. Use the results to discuss how increased atmospheric aerosol concentrations due to industrial emissions, which increases SW absorption, can mask the surface warming due to CO$_2$ increase.

## Solution 
The new energy balance equation for the surface is: 
\begin{equation}\frac{S_0}{4}(1-\Delta)(1-\alpha) +\epsilon \sigma T_a^4 = \sigma T^4\end{equation}

The new energy balanance equation for the atmosphere is: 
\begin{equation}\Delta\frac{S_0}{4} + \epsilon \sigma T^4 = 2\epsilon \sigma T_a^4 \end{equation}

where $T$ is the surface temperature, and $T_a$ is the atmospheric temperature. Substituting the second equation into the first yields:
\begin{equation} \frac{S_0}{4}(1-\Delta)(1-\alpha) +\Delta\frac{S_0}{8} + \frac{\epsilon \sigma T^4}{2} = \sigma T^4\end{equation}
Solving for $T$ yields:
\begin{equation}T = \left[\frac{\frac{S_0}{4}\left( (1-\Delta)(1-\alpha)+\frac{\Delta}{2}\right)}{\sigma(1-\frac{\epsilon}{2})} \right]^{1/4} \end{equation}

```python
# evaluating the solution numerically:

Delta = 0.15
T = (((S0/4)*((1-Delta)*(1-albedo)+Delta/2))/(sigma*(1-epsilon/2)))**(1/4)
print('T = ', T, "in celsius:",T-273.15)
T_a = (((Delta*S0/4) +epsilon*sigma*T**4) /(2*epsilon*sigma))**(1/4)
print('T_a = ',T_a)
```

(c) Consider an abrupt CO$_2$ doubling corresponding to the transition from panel a of Figure 2.4 to panel b. Considering the response of an atmosphere with a continuous vertical temperature profile of a prescribed lapse rate, calculate the increase in the emission height corresponding to an increase in radiative forcing by 4 W\,m$^{-2}$ assuming a lapse rate of 6.5 K/km.

```python
# first, the temperature decrease required to lower the LW by 4 W/m^2:

# The emission height temperature is found by the balance of outgoing 
# long-wave radiation from the emission height balancing the incoming solar 
# radiation minus the refelected part.
T=((S0/4)*(1-albedo)/sigma)**0.25
# calculate the value of DeltaT that leads to a 4W/m^2 warming
DeltaT=((4+sigma*T**4)/sigma)**0.25-T
print("DeltaT=",DeltaT)
# check that radiative forcing changed as expected:
print("DeltaF=",sigma*(T+DeltaT)**4-sigma*T**4)

# now, given a lapse rate of 6.5 K/km, this implies the following change in 
# emission height:
print("change in emission height (m):",(DeltaT/6.5)*1000)
```

## 3) Radiative forcing

  (a) Explain the upper and lower boundaries of the trapezoidal gap in the shape of the OLR curve around 600-800 cm$^{-1}$ (Figure 2.5b) due to CO$_2$ absorption, in terms of the emission height.
  
  (b) Plot the OLR in a standard atmosphere in the presence of 280 and 560 ppm CO2, ignoring all other greenhouse gasses.
  
  (c) Estimate the radiative forcing of 280 ppm of CO$_2$ by fitting a trapezoid to your plotted OLR curve and calculating its area.
  
  (d) Calculate the radiative forcing due to a doubling of the CO$_2$ from the preindustrial value of 280 ppm, by integrating over the difference of the two OLR curves.

## Solutions
a) The upper boundary of the trapezoid shape follows the surface blackbody curve, and is where the surface OLR curve would be if we did NOT have absorption by CO$_2$. Without absorption from greenhouse gases, all longwave radiation emitted from the surface would make it out to space, and thus we would say that the level of last absoprtion is at the surface. However, once we do enter a wavenumber range in which CO$_2$ can absorb, the level of last absorption gets pushed off the surface because radiation emitted from the surface at those wavenumbers is absorbed by CO$_2$ molecules directly above it. Due to the lapse rate (temperature is decreasing as you go up in the atmosphere), this means the level of last absorption is radiating at colder and colder temperatures. The flat bottom of the trapezoid occurs when your level of last absorption is pushed beyond the tropopause and into the stratosphere. If you assume an isothermal stratosphere, pushing the level of last absorption up even farther cannot decrease the OLR because you are still radiating at the same temperature. 

```python
# (b) 

# Read in pickle data of the two OLR curves and plot
fig = plt.figure(1,figsize=(6,4),dpi=300)
ax = plt.gca()
#plt.rcParams['figure.dpi']= 1200
#plt.rcParams.update({'font.size': 21})
plt.plot(wavenumbers,OLR_280ppm_CO2_only,lw=1,color="blue",label="280 ppm")
plt.plot(wavenumbers,OLR_560ppm_CO2_only,lw=1,color="red",ls='--',label="560 ppm")
plt.xlabel('Wavenumber (cm$^{-1}$)')
plt.ylabel('OLR (W/[m$^2$cm$^{-1}$])')

# plot tickmarks and grid
ax.xaxis.set_minor_locator(MultipleLocator(50))
ax.yaxis.set_minor_locator(MultipleLocator(.02))

plt.grid(which='both',lw=0.25)
plt.legend();
plt.show();
```

c) We choose the top of the trapezoid as halfway between the two points where the CO$_2$ band rejoins the surface blackbody curve at about .385 W/m$^2$ and the base at about .14 W/m$^2$. This gives a total height of the trapezoid as 0.245 W/m$^2$. The bottom width is about 75 cm$^{-1}$ and the top width is about 175 cm$^{-1}$. Thus using the formula for a trapezoid, $A = H \times (Base_1 + Base_2)/2$ we get a radiative forcing of  ~30.6 W/m$^2$.

d) The integral over the OLR curve tells you the total outgoing longwave radiation. The difference between the total OLR with 280ppm and 560ppm gives you the radiative forcing of doubling CO2.

```python
spect_res = wavenumbers[1] - wavenumbers[0]
RF = spect_res*np.sum(OLR_280ppm_CO2_only-OLR_560ppm_CO2_only)
print("Radiative Forcing: "+str(round(RF,3))+' W/m^2')
```

## 5) Global warming potential: 

Suppose the GWP of methane for a time horizon of 100 yr is 34. Assume methane concentration decays exponentially over 12.4 yr (so the decay function is $e^{-t/12.4}$) while that of CO$_2$ decays over 200 yr (e.g., due to dissolution in the ocean; note that while such a timescale is often used, the actual expected residence time of anthropogenic CO$_2$ in the climate system is thousands of years.) (a) How much more efficient is a kilogram of methane at absorbing LW radiation than a kilogram of CO$_2$? (b) Explain the difference between GWP and absorption efficiency. (c) What is the methane GWP for a time horizon of 500 yr? Explain.

### solution:

(a) We are given the Greenhouse Warming Potential and the decay functions, so we can reformulate the GWP equation to solve for $a_x/a_r$. Reformulate the equation as such:
\begin{equation} \frac{a_x}{a_r} = 34 *\frac{\int_{0}^{100} r(t)dt}{\int_{0}^{100}x(t)dt} \end{equation}

The decay functions for methane and CO$_2$ respectively are $x(t) = e^{-t/12.4}$ and $r(t) = e^{-t/200}$ where $t$ is the time in years. Calculating the definite integrals yields:
\begin{equation} \frac{a_x}{a_r} = 34 *\frac{\int_{0}^{100} e^{-t/200}dt}{\int_{0}^{100}e^{-t/12.4}dt} = 34 *\frac{78.694}{12.396} = 215.84 \end{equation}
Therefore, we can conclude that kg methane absorbs more than 200 times more strongly than a kg of CO$_2$.

(b) GWP takes into accout the residence time of a greenhouse gas, while absorption efficiency is a property per kg of the released gas regardless of how long it will be effective.

(c) The methane GWP over 500 years?

\begin{equation} GWP  = \frac{a_x}{a_r} * \frac{\int_{0}^{500} e^{-t/12.4}dt}{\int_{0}^{500}e^{-t/200}dt} \approx 14.5 \end{equation}

## 6) Water vapor feedback to increased CO$_2$: 
    
Modify the energy balance model equations (2.2) to include the effects of increasing water vapor and calculate the warming as a function of CO$_2$ with and without this feedback as follows. Assume the atmospheric emissivity depends on CO$_2$ and moisture $q(T_a)$ as $\epsilon({\mathrm{CO}}{{}_2},q(T_a)) =\epsilon_0+A\log_2({\mathrm{CO}}{{}_2}/280)+B\log_2({q(T_a)/q(T_{a0})})$, where $\epsilon_0=0.75$ and $T_{a0}$ is the atmospheric temperature for CO$_2$=280 ppm, and that the atmospheric relative humidity is the same before and after the CO$_2$ change. Find values for $A$ and $B$ such that the warming due to CO$_2$ doubling only is 2$^\circ$C, and due to both feedbacks is 4$^\circ$C. Plot $T_a({\mathrm{CO}}{{}_2})$ for CO$_2$ in the range of 200 to 800 ppm with and without the water vapor feedback.

### Solution:

The atmsopheric temperature is given by
\begin{align*}
T_a = \left(\frac{S_0(1-\alpha)}{4\sigma(2-\epsilon(CO_2,T_a))}\right)^{1/4} \\
\end{align*}
where the emissivity is written as a function of CO$_2$ and temperature. The dependence of temperature is due to the assumed dependence on moisture, which is a function of temperature.

At 280 ppm of CO$_2$, $\epsilon = \epsilon_0$ and $T_a = T_{a0} = 239.03$K. If we only consider the warming from CO$_2$ doubling, we have $\epsilon = \epsilon_0 + A$, and $A=0.041$ makes $T_a=241.03$ as desired, a 2K of warming.

Next, to find the $B$ coefficient in the emissivity formulation, we plot T_a for CO$_2=560$ ppm as a function of $B$ and find the value of $B$ that results in the temperature specified in the question (4 K warming).

Finally, now that we found $B$, we can use the above equation as an implicit equation for $T_a$ which needs to be solved for each value of CO$_2$.

```python
# constants:
albedo = 0.32
epsilon_0 = 0.75
Pa=500 # mid-atmospheric pressure 

# atmospheric temperature at 280 ppm:
T_a0 = ((S0/4)*(1-albedo)/(sigma*(2-epsilon_0)))**(1/4)

# first, find what changes to the emissivity are needed to cause the 
# suggested warmings:
# Try different values of A until the temperature rise due to CO2 doubling alone is 2 K:
A = 0.041  # change due to CO2 doubling
T_a_2XCO2 = ((S0/4)*(1-albedo)/(sigma*(2-(epsilon_0+A) )))**(1/4)
print("T_a0=",T_a0,", T_a_2XCO2=", T_a_2XCO2)
```

Now find B:

```python
# help functions: emissivity and saturation specific humidity:
def emissivity(CO2,T_a,A,B):
    # emisivity a function of CO2 and atmc temperature
    epsilon=epsilon_0+A*np.log2(CO2/280)+B*np.log2(q_sat(T_a,Pa)/q_sat(T_a0,Pa))
    return epsilon

def q_sat(T,P):
    # saturation specific humidity (g water vapor per g moist air):
    # inputs:
    # T: temperature, in Kelvin
    # P: pressure, in mb
    R_v = 461 # Gas constant for moist air = 461 J/(kg*K)
    R_d = 287 # Gas constant 287 J K^-1 kg^-1
    TT = T-273.15 # Kelvin to Celsius
    # Saturation water vapor pressure (mb) from Emanuel 4.4.14 p 116-117: 
    ew = 6.112*np.exp((17.67 * TT) / (TT + 243.5))
    # saturation mixing ratio (g water vapor per g dry air):
    rw = (R_d / R_v) * ew / (P - ew)
    # saturation specific humidity (g water vapor per g moist air):
    qw = rw / (1 + rw)
    return qw


# the desired temperature with both feedbacks as specified in the question:
T_a_2XCO2_water_vapor = T_a_2XCO2+2; 

# now calculate and plot the temperature as a function of the coefficient B to find 
# when it reaches the assumed warming:

Brange = np.arange(0.06, 0.08, 0.001)
Trange=0.0*Brange
i=-1
for B in Brange:
    i=i+1
    epsilon=emissivity(560,T_a_2XCO2_water_vapor,A,B)
    Trange[i]=((S0/4)*(1-albedo)/(sigma*(2-epsilon)))**(1/4)

fig = plt.figure(figsize=(6,3),dpi=300)
plt.plot(Brange, Trange, color = 'r' \
         , label='temperature with water vapor feedback')
plt.plot(Brange, Trange*0+T_a_2XCO2_water_vapor, color = 'b' \
         , label='T_a_2XCO2_water_vapor')
plt.legend()
plt.xlabel('B')
plt.ylabel('T_a')
plt.grid(lw=0.25)
plt.title('finding the water vaport feedback coeff');
plt.show();

```

Use A,B to solve for Ta and plot with and without the water vapor feedback:

```python
# from the plot, we conclude that
B=0.072
# results in the desired temperature.

# Now we can plot the temperature with and without the water vapor 
# feedback as function of CO2.
def solve_for_T_a(T_a,A,B):
    # calculate T_a-((S0/4)*(1-albedo)/(sigma*(2-epsilon(CO2,T_a))))**(1/4), 
    # to be called by fsolve in order to find a solution for T_a that satifies 
    # this equation
    equation=T_a-((S0/4)*(1-albedo)/(sigma*(2-emissivity(CO2,T_a,A,B))))**(1/4)
    return equation

# range of CO2 to consider:
CO2_range=np.arange(280,560,1)
# Prepare arrays for atmospheric temperature response to CO2 
# and to both CO2 and water vapor:
Ta_CO2_only=0.0*CO2_range
Ta_both=0.0*CO2_range

i=-1
for CO2 in CO2_range:
    i=i+1
    B=0; epsilon=emissivity(CO2,T_a0,A,B)
    Ta_CO2_only[i]=((S0/4)*(1-albedo)/(sigma*(2-epsilon)))**(1/4)
    B=0.072
    x0=240 # initial guess
    T_a=optimize.fsolve(solve_for_T_a,x0,args=(A,B))
    Ta_both[i]=T_a.item()


# Plot Ta vs CO2 with and without water vapor feedback:
fig = plt.figure(figsize=(6,3),dpi=300)
plt.plot(CO2_range, Ta_CO2_only, color = 'orange', label='CO$_2$ only')
plt.plot(CO2_range, Ta_both, color = 'red', label='CO$_2$ and water vapor')
plt.legend()
plt.xlabel('CO$_2$')
plt.ylabel('$T_a$')
plt.grid(lw=0.25)
plt.title('warming with and without water vaport feedback');
plt.show();
```
