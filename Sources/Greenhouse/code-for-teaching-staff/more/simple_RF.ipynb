```python
import numpy as np
import matplotlib.pyplot as plt

#constants 
l_k = 11.25 #cm^-1
k_0 = 671.0 #cm^-1
kap_0 = 40.0 #m^2/kg
c = 2.99792458*10**8 #m/s
kb = 1.38*10**-23 #J/K
h = 6.62607004*10**-34
Tsurf = 288 #K
Tstrat = 216 #K
psurf = 101325 #Pascal
ptrop = 19400 #Pascal
g = 9.81 #m/s^2

def plank(wavenum,t): #wavenum in cm^-1
    wavenum = wavenum*100
    return 100*(2*h*c**2*np.power(wavenum,3))*np.power(np.exp(h*c*wavenum/(kb*t))-1,-1)

def radiative_forcing(conc):
    conc = conc*1e-6
    return 2*l_k*np.log(g/(conc*psurf*ptrop*kap_0))*np.pi*(plank(k_0,Tsurf)-plank(k_0,Tstrat))

fig = plt.figure(figsize=(11,8))
plt.rcParams.update({'font.size': 18})
plt.xlabel('Concentration of CO2 (ppm)')
plt.ylabel('Radiative Forcing (W/m$^2$)')
concentrations = np.linspace(1,1000,10000)
rf = radiative_forcing(concentrations)
plt.plot(concentrations,rf)
plt.show();
```
